```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# Doc anh
im_path = 'C:/Users/ADMIN/Desktop/WorkSpace/Anh/02.jpg'
image = Image.open(im_path)

# Anh xam
gray_image = image.convert('L')

# Chuyen anh ve mang
image_a = np.array(gray_image)

# Tinh gradient the huong x
gradient_x = ndimage.convolve(image_a, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
# Tinh gradient the huong y
gradient_y = ndimage.convolve(image_a, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Tinh bien do gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Hien thi anh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Anh xam goc')

plt.subplot(2, 1, 2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('axis=0')
```

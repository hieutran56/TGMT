```python
from PIL import Image
from imgtools import *
```

```python
my_path = 'C:/Users/ADMIN/Desktop/WorkSpace/Anh/02.jpg'
```

```python
img = load_image(my_path)
```

```python
display(img)
```

```python
#Định vị khu vực cắt
region=(50,50,200,200)
#Cắt vùng ảnh
cropper_image=img.crop(region)
display(cropper_image)
```

```python
#Dán hình
#Xác định vị trí cần dán
paste_poition = (250,200)
img.paste(cropper_image,paste_poition)
display(img)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
my_path = 'C:/Users/ADMIN/Desktop/WorkSpace/Anh/02.jpg'
im = Image.open(my_path)
plt.imshow(im)
```

```python
# Chuyen thanh mau xam
im_gray=im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array=np.array(im_gray)
print(im_array)
```

```python
# Tao 1 hinh ve moi
plt.figure()
plt.hist(im_array.flatten(),bins=128)
plt.title('Bieu do Histogram cua anh')
plt.xlabel('Gia tri cua pixel')
plt.ylabel('So luong')
plt.show()
```

```python
#Lam noi bat duong bien
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

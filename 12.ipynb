```python
my_path = 'C:/Users/ADMIN/Desktop/WorkSpace/Anh/02.jpg'

#Goi anh
im=Image.open(my_path)

#Chuyen hinh ve mang numpy
im_array=np.array(im.convert("L"))

#Bien doi thanh anh xam theo yeu cau
inverted_image = 255-im_array #Inverted anh bang cach lay 255 - gia tri pixel
squared_image = im_array**2 # Binh phuong gia tri pixel
clamped_image = np.clip(im_array, 200, 400) #Gioi han gia tri pixel trong khoang 200, 400

# Hien thi hnh anh goc va hinh anh da bien doi
fig, axs =plt.subplots(2, 2, figsize=(10,10))
fig.subplots_adjust(wspace=0.2, hspace=0.2)

axs[0, 0].imshow(im_array, cmap="gray")
axs[0,0].set_title('Hinh anh goc')

axs[0, 1].imshow(inverted_image, cmap="gray")
axs[0, 1].set_title('Hinh anh da bien doi')

axs[1, 0].imshow(squared_image, cmap="gray")
axs[1, 0].set_title('Hinh anh da bien doi')

axs[1, 1].imshow(clamped_image, cmap="gray")
axs[1, 1].set_title('Hinh anh da bien doi')

plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# Doc anh
im_path = 'C:/Users/ADMIN/Desktop/WorkSpace/Anh/02.jpg'
image = Image.open(im_path)

# Anh xam
gray_image = image.convert('L')

# Chuyen anh ve mang
image_a = np.array(gray_image)

#Lam mo anh xam ve mang
blurred_image_a = ndimage.gaussian_filter(image_a, sigma=0.5, mode="constant")
blurred_image_b = ndimage.gaussian_filter(image_a, sigma=1, mode="constant")
blurred_image_c = ndimage.gaussian_filter(image_a, sigma=10, mode="constant")

#Hien thi anh
plt.subplot(2, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Anh xam goc')

plt.subplot(2, 2, 2)
plt.imshow(blurred_image_a, cmap='gray')
plt.title('sigma=0.5')

plt.subplot(2, 2, 3)
plt.imshow(blurred_image_b, cmap='gray')
plt.title('sigma=1')

plt.subplot(2, 2, 4)
plt.imshow(blurred_image_c, cmap='gray')
plt.title('sigma=10')
```
